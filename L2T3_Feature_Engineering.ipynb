```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Dataset .csv")
print(df.columns)
print(df.head())
```

```python
# Length of restaurant name
df["Restaurant Name Length"] = df["Restaurant Name"].astype(str).apply(len)

# Length of address
df["Address Length"] = df["Address"].astype(str).apply(len)

# Number of words in restaurant name
df["Restaurant Name Word Count"] = df["Restaurant Name"].astype(str).apply(lambda x: len(x.split()))

# Number of cuisines offered
df["Number of Cuisines"] = df["Cuisines"].astype(str).apply(lambda x: len(x.split(", ")))
```

```python
df["Has Table Booking"] = df["Has Table booking"].map({"Yes": 1, "No": 0})
df["Has Online Delivery"] = df["Has Online delivery"].map({"Yes": 1, "No": 0})
```

```python
# Convert 'City' into multiple binary columns
city_encoded = pd.get_dummies(df["City"], prefix="City")

# Merge back to main dataframe
df = pd.concat([df, city_encoded], axis=1)
```

```python
# Convert numeric price range into labels
price_map = {
    1: "Low",
    2: "Medium",
    3: "High",
    4: "Very High"
}

df["Price Category"] = df["Price range"].map(price_map)
```

```python
print("\nNew Feature Columns:\n")
print(df[[
    "Restaurant Name Length",
    "Address Length",
    "Restaurant Name Word Count",
    "Number of Cuisines",
    "Has Table Booking",
    "Has Online Delivery",
    "Price Category"
]].head())
```

```python
df.to_csv("feature_engineered_dataset.csv", index=False)

print("\nFeature engineered dataset saved successfully!")
```
